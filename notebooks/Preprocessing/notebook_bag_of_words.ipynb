```python
import re
import pandas as pd
import numpy as np

from sklearn.feature_extraction.text import CountVectorizer
```

```python
df = pd.read_csv(r'C:\Users\Checkout\Downloads\cyberbullying-classification-master\data\train_data.csv')

# printing the shape: (rows, columns)
display(df.shape)

# first 5 datapoints
display(df.head())

# 5 random samples
display(df.sample(5))
```

```python
cv = CountVectorizer()
counts = cv.fit_transform(df['clean_tweet'].values.astype('U'))

```

```python
df_bow = pd.DataFrame(counts.A, columns=cv.get_feature_names())
df_bow['cyberbullying_type'] = df['cyberbullying_type']
```
